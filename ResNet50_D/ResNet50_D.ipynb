```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet50_D.txt','r')as f:
    acc_ResNet50_D = f.readlines()
    acc_ResNet50_D = [i.replace('\n','') for i in acc_ResNet50_D]
    acc_ResNet50_D = np.array(acc_ResNet50_D,dtype=float)

plt.figure()
plt.plot(acc_ResNet50_D,label='ResNet-50-D')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
