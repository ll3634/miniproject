```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet18_B.txt','r')as f:
    acc_ResNet18_B = f.readlines()
    acc_ResNet18_B = [i.replace('\n','') for i in acc_ResNet18_B]
    acc_ResNet18_B = np.array(acc_ResNet18_B,dtype=float)

plt.figure()
plt.plot(acc_ResNet18_B,label='ResNet-18-B')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
