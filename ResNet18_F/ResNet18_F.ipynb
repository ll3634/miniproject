```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet18_F.txt','r')as f:
    acc_ResNet18_F = f.readlines()
    acc_ResNet18_F = [i.replace('\n','') for i in acc_ResNet18_F]
    acc_ResNet18_F = np.array(acc_ResNet18_F,dtype=float)

plt.figure()
plt.plot(acc_ResNet18_F,label='ResNet-18-F')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
