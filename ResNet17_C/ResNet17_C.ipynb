```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet17_C.txt','r')as f:
    acc_ResNet17_C = f.readlines()
    acc_ResNet17_C = [i.replace('\n','') for i in acc_ResNet17_C]
    acc_ResNet17_C = np.array(acc_ResNet17_C,dtype=float)

plt.figure()
plt.plot(acc_ResNet17_C,label='ResNet-17-C')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
