```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet50_F.txt','r')as f:
    acc_ResNet50 = f.readlines()
    acc_ResNet50 = [i.replace('\n','') for i in acc_ResNet50]
    acc_ResNet50 = np.array(acc_ResNet50,dtype=float)

plt.figure()
plt.plot(acc_ResNet50,label='ResNet-50-F')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
