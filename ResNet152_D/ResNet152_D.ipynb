```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet152_D.txt','r')as f:
    acc_ResNet152_D = f.readlines()
    acc_ResNet152_D = [i.replace('\n','') for i in acc_ResNet152_D]
    acc_ResNet152_D = np.array(acc_ResNet152_D,dtype=float)

plt.figure()
plt.plot(acc_ResNet152_D,label='ResNet-152-D')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
