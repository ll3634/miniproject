```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet18_D.txt','r')as f:
    acc_ResNet18_D = f.readlines()
    acc_ResNet18_D = [i.replace('\n','') for i in acc_ResNet18_D]
    acc_ResNet18_D = np.array(acc_ResNet18_D,dtype=float)

plt.figure()
plt.plot(acc_ResNet18_D,label='ResNet-18-D')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
