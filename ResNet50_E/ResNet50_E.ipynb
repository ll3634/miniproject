```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet50_E.txt','r')as f:
    acc_ResNet50_E = f.readlines()
    acc_ResNet50_E = [i.replace('\n','') for i in acc_ResNet50_E]
    acc_ResNet50_E = np.array(acc_ResNet50_E,dtype=float)

plt.figure()
plt.plot(acc_ResNet50_E,label='ResNet-50-E')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
