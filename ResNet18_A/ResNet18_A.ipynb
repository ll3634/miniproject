```python
import matplotlib.pyplot as plt
import numpy as np

with open('acc_ResNet18_A.txt','r')as f:
    acc_ResNet18_A = f.readlines()
    acc_ResNet18_A = [i.replace('\n','') for i in acc_ResNet18_A]
    acc_ResNet18_A = np.array(acc_ResNet18_A,dtype=float)

plt.figure()
plt.plot(acc_ResNet18_A,label='ResNet-18-A')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
